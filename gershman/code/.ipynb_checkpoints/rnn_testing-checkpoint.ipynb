```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

import time
import random
from sklearn.preprocessing import MinMaxScaler

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset

import warnings
warnings.filterwarnings('ignore')

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
class GRU_RNN_TWO(nn.Module):
        
    def __init__(self,input_size, hidden_size, num_beta_embedding, num_gamma_embedding, output_size, dropout):
        
        super(GRU_RNN_TWO, self).__init__()
        
        self.input_size = input_size
        self.num_beta_embedding = num_beta_embedding
        self.num_gamma_embedding = num_gamma_embedding
        
        
        self.hidden_0 = nn.GRU(  
                    input_size=input_size,
                    hidden_size=hidden_size,
                    num_layers=1,
                    batch_first=True,
                    dropout=dropout
        )
        
        self.hidden_1 = nn.GRU(  
                    input_size=input_size + num_beta_embedding + num_gamma_embedding,
                    hidden_size=hidden_size,
                    num_layers=1,
                    batch_first=True,
                    dropout=dropout 
        )
    
    
        self.out_beta = nn.Linear(hidden_size, num_beta_embedding)
        self.out_gamma = nn.Linear(hidden_size, num_gamma_embedding)
        
        self.relu_beta = nn.ReLU()
        self.relu_gamma = nn.ReLU()

        self.reg_beta = nn.Linear(num_beta_embedding, 1)
        self.reg_gamma = nn.Linear(num_gamma_embedding, 1)
        
        self.out_action = nn.Linear(hidden_size, output_size)
        
    def forward(self, x):

        # parameters estimation
        output_0, hn_0 = self.hidden_0(x)
        
        output_beta = self.out_beta(output_0)
        output_beta = self.relu_beta(output_beta)
        
        output_gamma = self.out_gamma(output_0)
        output_gamma = self.relu_gamma(output_gamma)
        
        cont_output_beta = self.reg_beta(output_beta)
        cont_output_gamma = self.reg_gamma(output_gamma)
    
        # concat input 
        input_1 = torch.cat([x[0],output_beta[0],output_gamma[0]],dim=1)
        input_1 = input_1.reshape(1,x.shape[1],self.input_size + self.num_beta_embedding + self.num_gamma_embedding)
        
        # action predication
        output_action, hn_1 = self.hidden_1(input_1)
        output_action = self.out_action(output_action)
        output_action = F.softmax(output_action,dim=-1)

        return output_beta, output_gamma, cont_output_beta, cont_output_gamma, output_action, hn_0, hn_1
    
```

```python
class behavior_dataset(Dataset):
    """ 
    """
    def __init__(self,dataframe):
        
        # action one hot transformation 
        action = np.array(dataframe['action'])
        if np.all(action == action[0]):
            action = np.append(action,(1-action[0]))
            action = torch.tensor((action).reshape(len(dataframe) + 1),dtype=int)
            action_onehot = nn.functional.one_hot(action, len(action.unique()))
            # delete last one
            action_onehot = action_onehot[:-1]
        else:
            action = torch.tensor((action).reshape(len(dataframe)),dtype=int)
            action_onehot = nn.functional.one_hot(action, len(action.unique()))
        
        # reward
        reward = torch.tensor((np.array(dataframe['reward'])).reshape(len(dataframe)),dtype=torch.float32)
        
        # concatinating reward and action
        reward_action = torch.cat([reward[ :, np.newaxis], action_onehot],1)
        
        # adding dummy zeros to the beginning and ignoring the last one
        reward_action_shift = nn.functional.pad(reward_action,[0,0,1,0])[:-1]
        
        # network input 
        x = reward_action_shift
        n_blocks = int(len(dataframe)/10)
        x.reshape(n_blocks,10,INPUT_SIZE)[:,0,:] = torch.zeros(size=(n_blocks,INPUT_SIZE))
        
        # network output 
        y = action_onehot
  
        self.x = x.type(dtype=torch.float32)
        self.y = y.type(dtype=torch.float32)
        self.len = len(dataframe)

    def __getitem__(self,idx):
        return self.x[idx],self.y[idx]
  
    def __len__(self):
        return self.len    
    
    
class merge_behavior_dataset(Dataset):
    """ 
    Merge Dataset of each agent to one dataset
    
    Args:
        dataset_list: list of Dataset of all agent 
        n_trials: num_of_trials each agent was simulated
        
    Returns: 
        torch Dataset:
        x: [reward_(t-1) , action_(t-1)] all agents
        y: [action_t, parameter embedding] all agents
        
    """
    def __init__(self, dataset_list, n_trials):
        X = []
        Y = []
        for j,agent in enumerate(dataset_list):
            for i in range(n_trials[j]):
                X.append(agent[i][0])
                Y.append(agent[i][1])
                
        self.x = torch.stack(X).type(dtype=torch.float32)
        self.y = torch.stack(Y).type(dtype=torch.float32)
        self.len = len(X)   
    
    def __getitem__(self,idx):
        return self.x[idx],self.y[idx]
  
    def __len__(self):
        return self.len    
```

```python
load = torch.load(f'checkpoint_best_val.pth',map_location='cpu')

INPUT_SIZE = 3
aHidden = 32
num_beta_embedding = 4
num_gamma_embedding = 4 
OUTPUT_SIZE = 2

rnn = GRU_RNN_TWO(
              input_size=INPUT_SIZE,
              hidden_size=aHidden,
              num_beta_embedding=num_beta_embedding,
              num_gamma_embedding=num_gamma_embedding,
              output_size=OUTPUT_SIZE,
              dropout=0
            )

rnn.load_state_dict(load['model_state'])
print(load['epoch'])
rnn.eval()
```

```python
def bce_loss(y_hat,y_true):
    eps = 1e-10
    return (-((y_true*(np.log(y_hat + eps)) + (1-y_true)*(np.log(1-y_hat + eps)))))

df = pd.read_csv('../data/data2.csv')
df['action'] = df['choice']-1
df['block'] = df['block']-1
df['trial'] = df['trial']-1
df['subject'] = df['subject']-1


num_of_action = df['action'].nunique()
num_of_trials = df['trial'].nunique()*df['block'].nunique()
scaler = MinMaxScaler(feature_range=(0,1))
df['reward'] = scaler.fit_transform(df.reward.values.reshape(len(df),1)).flatten()

test_data = []

for i in range(44):
    cur_df = df[(df['subject']==i)].reset_index()
    test_data.append(cur_df)
    
all_data_rnn = []

for i in range(44):
    cur_df = test_data[i]
    cur_df = cur_df.reset_index()
    all_data_rnn.append(behavior_dataset(cur_df))
    
val_dataset = merge_behavior_dataset(all_data_rnn, np.repeat(200,44))
```

```python
predict_beta_array = []
predict_gamma_array = []
predict_p_0_array = []

j = 0

for i in range(44):
        
    rnn.to(device)
    cur_df = test_data[i]
    cur_df = cur_df.reset_index()
        
    y_hat_dis_beta, y_hat_dis_gamma, y_hat_cont_beta, y_hat_cont_gamma, y_hat_action, _, _ = rnn((val_dataset[j:200+j][0].reshape(1,num_of_trials,3)).to(device))
    y_hat_action = (y_hat_action.view(-1, num_of_action)) 
        
    predict_beta_array.append((y_hat_cont_beta[0].cpu().detach().numpy().flatten()))
    predict_gamma_array.append((y_hat_cont_gamma[0].cpu().detach().numpy().flatten()))
    predict_p_0_array.append(((y_hat_action[:,0]).cpu().detach().numpy()))
    
    j += 200
    
arr_bce = []
arr_psr = []
for i in range(44):
    bce = bce_loss(1-predict_p_0_array[i],test_data[i]['action'])
    arr_bce.append(bce.mean())
    arr_psr.append(1-(np.array(bce.sum()) / (-len(test_data[i])*np.log(0.5))))
    
bce = np.array(arr_bce)
psr = np.array(arr_psr)
    
print('BCE',bce.mean())
print('PSR',psr.mean())   

df = pd.read_csv('../results/fit_ger_loo.csv')

sns.scatterplot(bce,df.bce_test)
plt.plot([0.1,1],[0.1,1],'--k')
sns.despine()

# to beat 0.3833
```
