```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from scipy.stats import pearsonr

from q_sim import *
from q_fit import *
from q_sim_rw import *
from utils import *
```

```python
# config num of agent to simulate
num_of_agents = 300
# config num of trails for each agent
num_of_trials = 1_000
n_trials_per_block = 100

all_data = [] 

# stationarity
for agent in range(100):
        reward_probs = create_reward_probs(2,n_trials_per_block,num_of_trials,np.array([0.1,0.5,0.9]))
        param = configuration_parameters()

        df = q_sim(
                agent,
                param,
                num_of_trials,
                reward_probs,
                probability_to_switch_parameters=0.000,
                max_change=np.random.choice([2,3,4])
        )
        all_data.append(df)      

# non-stationarity
for agent in range(100,200):
        reward_probs = create_reward_probs(2,n_trials_per_block,num_of_trials,np.array([0.1,0.5,0.9]))
        param = configuration_parameters()
        df = q_sim(
                agent,
                param,
                num_of_trials,
                reward_probs,
                probability_to_switch_parameters=0.005,
                max_change= 4 # np.random.choice([2,3,4])
        )
        all_data.append(df)    
        
# random walk 
for agent in range(200,300):
        reward_probs = create_reward_probs(2,n_trials_per_block,num_of_trials,np.array([0.1,0.5,0.9]))
        param = configuration_parameters()
        df = q_sim_rw(
                agent,
                param,
                num_of_trials,
                reward_probs,
                [0.01,0.05]
        )
        all_data.append(df)  
        
df_all_agent = pd.concat(all_data)    
df_all_agent = df_all_agent.reset_index()
df_all_agent = df_all_agent.drop(columns=['index'])  

df_all_agent.to_csv(f'../data/{num_of_agents}_artificial_test_agents.csv',index=False)
```

```python
for i in range(200,300):
    fig, (ax0,ax1) = plt.subplots(1,2,figsize=(15,3))
    ax0.plot(all_data[i].alpha)
    ax1.plot(all_data[i].beta)
    plt.show()
```
